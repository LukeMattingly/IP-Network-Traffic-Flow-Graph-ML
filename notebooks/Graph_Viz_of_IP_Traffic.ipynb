```python

sample = data.head(5000)
```

```python
sample['Source.IP'].values
```

```python
net = NetworkViz(notebook=True)
net.add_nodes(sample['Source.IP'].values, label=sample['Source.IP'].values)
net.add_nodes(sample['Destination.IP'].values, label=sample['Destination.IP'].values)
net.show('ip_nodes.html')
```
